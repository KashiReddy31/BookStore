```python
#1
def categorize_numbers(numbers_list):
    counts = {
        "positive": 0,
        "negative": 0,
        "zero": 0
    }

    for num in numbers_list:
        if num > 0:
            counts["positive"] += 1
        elif num < 0:
            counts["negative"] += 1
        else:
            counts["zero"] += 1

    return counts

example_list = [1, -2, 0, 5, -10, 0, 7, -3]
result = categorize_numbers(example_list)
print(result)

example_list_2 = [10, 20, 30]
result_2 = categorize_numbers(example_list_2)
print(result_2)

example_list_3 = [-1, -5, -8]
result_3 = categorize_numbers(example_list_3)
print(result_3)

example_list_4 = [0, 0, 0]
result_4 = categorize_numbers(example_list_4)
print(result_4)
```

```python
#2
def compare_lists(list1, list2):

    if len(list1) != len(list2):
        raise ValueError("Both lists must be of equal length.")

    overall_max_value = float('-inf')
    overall_max_source_list = None
    list1_higher_count = 0
    list2_higher_count = 0

    for i in range(len(list1)):
        val1 = list1[i]
        val2 = list2[i]

        if val1 > overall_max_value:
            overall_max_value = val1
            overall_max_source_list = 'list1'
        if val2 > overall_max_value:
            overall_max_value = val2
            overall_max_source_list = 'list2'

        if val1 > val2:
            list1_higher_count += 1
        elif val2 > val1:
            list2_higher_count += 1

    return {
        'overall_max_value': overall_max_value,
        'overall_max_source_list': overall_max_source_list,
        'list1_higher_count': list1_higher_count,
        'list2_higher_count': list2_higher_count
    }

list_a = [10, 5, 20, 15, 30]
list_b = [8, 12, 18, 25, 28]
result_ab = compare_lists(list_a, list_b)
print(f"Result for list_a and list_b: {result_ab}")

list_c = [1, 2, 3]
list_d = [3, 2, 1]
result_cd = compare_lists(list_c, list_d)
print(f"Result for list_c and list_d: {result_cd}")

list_e = [5, 5, 5]
list_f = [5, 5, 5]
result_ef = compare_lists(list_e, list_f)
print(f"Result for list_e and list_f: {result_ef}")

list_g = [100, 10, 50]
list_h = [90, 200, 40]
result_gh = compare_lists(list_g, list_h)
print(f"Result for list_g and list_h: {result_gh}")

```

```python
#3
def is_leap_year(year):

    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def get_leap_years_comprehension(years_list):

    return [year for year in years_list if is_leap_year(year)]

years_1 = [2000, 2004, 1900, 2020, 2021, 2400]
leap_years_1 = get_leap_years_comprehension(years_1)
print(f"Original years: {years_1}")
print(f"Leap years (comprehension): {leap_years_1}")

years_2 = [1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004]
leap_years_2 = get_leap_years_comprehension(years_2)
print(f"\nOriginal years: {years_2}")
print(f"Leap years (comprehension): {leap_years_2}")

years_3 = [1800, 1900, 2100, 2200, 2300]
leap_years_3 = get_leap_years_comprehension(years_3)
print(f"\nOriginal years: {years_3}")
print(f"Leap years (comprehension): {leap_years_3}")

years_4 = [2024, 2028, 2032]
leap_years_4 = get_leap_years_comprehension(years_4)
print(f"\nOriginal years: {years_4}")
print(f"Leap years (comprehension): {leap_years_4}")
```

```python
#4
def fibonacci_dp(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]

    fib_values = [0] * n
    fib_values[0] = 0
    fib_values[1] = 1

    for i in range(2, n):
        fib_values[i] = fib_values[i-1] + fib_values[i-2]

    return fib_values

print(f"Fibonacci (DP) for n=10: {fibonacci_dp(10)}")
print(f"Fibonacci (DP) for n=1: {fibonacci_dp(1)}")
print(f"Fibonacci (DP) for n=0: {fibonacci_dp(0)}")
```

```python
#4
def fibonacci_naive_recursive(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_naive_recursive(n-1) + fibonacci_naive_recursive(n-2)

def get_fibonacci_list_naive(n):
    if n <= 0:
        return []
    return [fibonacci_naive_recursive(i) for i in range(n)]

print(f"Fibonacci (Naive Recursive) for n=10: {get_fibonacci_list_naive(10)}")
```

```python
#5
def calculate_factorial(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")
    if n < 0:
        raise ValueError("Input must be a non-negative integer.")
    if n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

print(f"Factorial of 5: {calculate_factorial(5)}")
print(f"Factorial of 0: {calculate_factorial(0)}")
print(f"Factorial of 1: {calculate_factorial(1)}")
print(f"Factorial of 10: {calculate_factorial(10)}")

print("\nTesting error handling:")
try:
    calculate_factorial(-5)
except ValueError as e:
    print(f"Caught expected error for -5: {e}")

try:
    calculate_factorial(3.14)
except TypeError as e:
    print(f"Caught expected error for 3.14: {e}")

try:
    calculate_factorial("abc")
except TypeError as e:
    print(f"Caught expected error for 'abc': {e}")
```

```python
#7
def sieve_of_eratosthenes(n):
    if not isinstance(n, int):
        raise TypeError("Input 'n' must be an integer.")
    if n < 0:
        raise ValueError("Input 'n' must be a non-negative integer.")
    if n < 2:
        return []
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False

    p = 2
    while (p * p <= n):

        if is_prime[p]:

            for i in range(p * p, n + 1, p):
                is_prime[i] = False
        p += 1

    primes = [p for p in range(n + 1) if is_prime[p]]
    return primes

print(f"Primes up to 10: {sieve_of_eratosthenes(10)}")
print(f"Primes up to 20: {sieve_of_eratosthenes(20)}")
print(f"Primes up to 2: {sieve_of_eratosthenes(2)}")
print(f"Primes up to 0: {sieve_of_eratosthenes(0)}")
print(f"Primes up to 1: {sieve_of_eratosthenes(1)}")
print(f"Primes up to 50: {sieve_of_eratosthenes(50)}")

```

```python
#8
def calculate_summary_statistics(numbers_list):

    if not isinstance(numbers_list, list):
        raise TypeError("Input must be a list.")

    if not numbers_list:

        return {
            'mean': None,
            'median': None,
            'min': None,
            'max': None,
            'sum': None
        }


    for item in numbers_list:
        if not isinstance(item, (int, float)):
            raise TypeError("All elements in the list must be numbers (int or float).")

    total_sum = sum(numbers_list)
    minimum = min(numbers_list)
    maximum = max(numbers_list)
    count = len(numbers_list)

    mean_value = total_sum / count

    sorted_list = sorted(numbers_list)
    mid = count // 2
    if count % 2 == 0:
        median_value = (sorted_list[mid - 1] + sorted_list[mid]) / 2
    else:
        median_value = sorted_list[mid]

    return {
        'mean': mean_value,
        'median': median_value,
        'min': minimum,
        'max': maximum,
        'sum': total_sum
    }


list_odd = [1, 5, 2, 8, 3]
print(f"List: {list_odd}")
print(f"Statistics: {calculate_summary_statistics(list_odd)}")

list_even = [10, 20, 15, 25]
print(f"\nList: {list_even}")
print(f"Statistics: {calculate_summary_statistics(list_even)}")

list_floats = [1.1, 2.2, 3.3]
print(f"\nList: {list_floats}")
print(f"Statistics: {calculate_summary_statistics(list_floats)}")

empty_list = []
print(f"\nList: {empty_list}")
print(f"Statistics: {calculate_summary_statistics(empty_list)}")

single_element_list = [42]
print(f"\nList: {single_element_list}")
print(f"Statistics: {calculate_summary_statistics(single_element_list)}")


print("\nTesting error handling for non-numeric elements:")
try:
    calculate_summary_statistics([1, 2, 'a', 4])
except TypeError as e:
    print(f"Caught expected error: {e}")

try:
    calculate_summary_statistics("not a list")
except TypeError as e:
    print(f"Caught expected error: {e}")
```

```python
#9
def count_chars(input_string):
    vowels = "aeiou"
    consonants = "bcdfghjklmnpqrstvwxyz"

    counts = {
        "vowels": 0,
        "consonants": 0,
        "others": 0
    }

    for char in input_string:
        char_lower = char.lower()

        if char_lower in vowels:
            counts["vowels"] += 1
        elif char_lower in consonants:
            counts["consonants"] += 1
        else:
            counts["others"] += 1

    return counts

string1 = "Hello World!"
print(f"String: '{string1}'")
print(f"Counts: {count_chars(string1)}")

string2 = "Python Programming 123."
print(f"\nString: '{string2}'")
print(f"Counts: {count_chars(string2)}")

string3 = "A B C D E F G"
print(f"\nString: '{string3}'")
print(f"Counts: {count_chars(string3)}")

string4 = ""
print(f"\nString: '{string4}'")
print(f"Counts: {count_chars(string4)}")

string5 = "?!@#$%^&*()_+"
print(f"\nString: '{string5}'")
print(f"Counts: {count_chars(string5)}")
```

```python
#10
import math

def calculate_circle_areas(radii_list):
    if not isinstance(radii_list, list):
        raise TypeError("Input must be a list of radii.")

    areas = []
    for radius in radii_list:
        if not isinstance(radius, (int, float)):
            raise TypeError("All elements in the radii list must be numbers.")
        if radius < 0:
            raise ValueError("Radius cannot be negative.")
        areas.append(math.pi * (radius ** 2))
    return areas

radii1 = [1, 2, 3, 4]
areas1 = calculate_circle_areas(radii1)
print(f"Radii: {radii1}, Areas: {areas1}")

radii2 = [0.5, 1.5, 2.5]
areas2 = calculate_circle_areas(radii2)
print(f"\nRadii: {radii2}, Areas: {areas2}")

radii3 = [0]
areas3 = calculate_circle_areas(radii3)
print(f"\nRadii: {radii3}, Areas: {areas3}")

empty_radii = []
empty_areas = calculate_circle_areas(empty_radii)
print(f"\nRadii: {empty_radii}, Areas: {empty_areas}")
```

```python
#13
def gcd_euclidean(a, b):
    if not isinstance(a, int) or not isinstance(b, int):
        raise TypeError("Both inputs must be integers.")
    if a < 0 or b < 0:
        raise ValueError("Both inputs must be non-negative integers.")

    while b:
        a, b = b, a % b
    return a

print(f"GCD(48, 18): {gcd_euclidean(48, 18)}")
print(f"GCD(101, 103): {gcd_euclidean(101, 103)}")
print(f"GCD(0, 5): {gcd_euclidean(0, 5)}")
print(f"GCD(15, 0): {gcd_euclidean(15, 0)}")
```

```python
#13
import itertools

def gcd_of_all_pairs_in_list(numbers_list):

    if not isinstance(numbers_list, list):
        raise TypeError("Input must be a list.")
    if not numbers_list:
        return []

    results = []
    for num1, num2 in itertools.combinations(numbers_list, 2):
        current_gcd = gcd_euclidean(num1, num2)
        results.append((num1, num2, current_gcd))
    return results

list_a = [12, 18, 24]
print(f"\nList: {list_a}")
print(f"GCD of all pairs: {gcd_of_all_pairs_in_list(list_a)}")

list_b = [7, 13, 29]
print(f"\nList: {list_b}")
print(f"GCD of all pairs: {gcd_of_all_pairs_in_list(list_b)}")

list_c = [10, 20, 30, 40]
print(f"\nList: {list_c}")
print(f"GCD of all pairs: {gcd_of_all_pairs_in_list(list_c)}")

empty_list = []
print(f"\nList: {empty_list}")
print(f"GCD of all pairs: {gcd_of_all_pairs_in_list(empty_list)}")
```

```python
#14
from collections import defaultdict
import re

def _clean_and_sort_for_key(s):

    return tuple(sorted(re.sub(r'[^a-z0-9]', '', s.lower())))

def group_anagrams(words_list):

    if not isinstance(words_list, list):
        raise TypeError("Input must be a list of words.")

    anagram_groups = defaultdict(list)
    for word in words_list:
        if not isinstance(word, str):
            raise TypeError(f"All elements in the list must be strings. Found type {type(word)} for value {word}.")
        cleaned_key = _clean_and_sort_for_key(word)
        anagram_groups[cleaned_key].append(word)

    return list(anagram_groups.values())

words1 = ["listen", "silent", "hello", "ehllo", "elbow", "below", "world", "tar", "rat", "art"]
print(f"Words: {words1}")
print(f"Grouped Anagrams: {group_anagrams(words1)}")

words2 = ["act", "cat", "tac", "dog", "god", "good"]
print(f"\nWords: {words2}")
print(f"Grouped Anagrams: {group_anagrams(words2)}")

words3 = []
print(f"\nWords: {words3}")
print(f"Grouped Anagrams: {group_anagrams(words3)}")

words4 = ["A man, a plan, a canal: Panama", "Panama, a canal: a plan, a man"]
print(f"\nWords: {words4}")
print(f"Grouped Anagrams: {group_anagrams(words4)}")

```

```python
#15
def factorial_iterative(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")
    if n < 0:
        raise ValueError("Input must be a non-negative integer.")

    if n == 0:
        return 1

    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(f"Iterative Factorial of 5: {factorial_iterative(5)}")
print(f"Iterative Factorial of 0: {factorial_iterative(0)}")
print(f"Iterative Factorial of 10: {factorial_iterative(10)}")
```

```python
#15
def factorial_recursive(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")
    if n < 0:
        raise ValueError("Input must be a non-negative integer.")
    if n == 0:
        return 1
    else:
        return n * factorial_recursive(n - 1)

print(f"Recursive Factorial of 5: {factorial_recursive(5)}")
print(f"Recursive Factorial of 0: {factorial_recursive(0)}")
print(f"Recursive Factorial of 10: {factorial_recursive(10)}")
```

```python
#16
import math

def is_prime_single(num):
    if num < 2:
        return False
    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

def is_perfect_square(num):
    if num < 0:
        return False
    sqrt_num = int(math.sqrt(num))
    return sqrt_num * sqrt_num == num

def classify_integer_properties(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")

    properties = {
        "prime": is_prime_single(n),
        "perfect_square": is_perfect_square(n),
        "even_odd": "",
        "sign": ""
    }


    if n % 2 == 0:
        properties["even_odd"] = "even"
    else:
        properties["even_odd"] = "odd"

    if n > 0:
        properties["sign"] = "positive"
    elif n < 0:
        properties["sign"] = "negative"
    else:
        properties["sign"] = "zero"

    return properties

print(f"Properties of 7: {classify_integer_properties(7)}")
print(f"Properties of 16: {classify_integer_properties(16)}")
print(f"Properties of 10: {classify_integer_properties(10)}")
print(f"Properties of 0: {classify_integer_properties(0)}")
print(f"Properties of -5: {classify_integer_properties(-5)}")
print(f"Properties of 1: {classify_integer_properties(1)}")
print(f"Properties of 2: {classify_integer_properties(2)}")
print(f"Properties of 9: {classify_integer_properties(9)}")
print(f"Properties of 101: {classify_integer_properties(101)}")
```

```python
#17
import re

def is_palindrome(text):
    cleaned_text = re.sub(r'[^a-z0-9]', '', text.lower())
    return cleaned_text == cleaned_text[::-1]

def analyze_sentence(sentence_string):

    if not isinstance(sentence_string, str):
        raise TypeError("Input must be a string.")

    analysis = {
        "word_count": 0,
        "unique_word_count": 0,
        "longest_word": "",
        "total_vowel_count": 0,
        "is_palindrome": False
    }

    words = re.findall(r'\b[a-zA-Z]+\b', sentence_string.lower())
    analysis["word_count"] = len(words)
    analysis["unique_word_count"] = len(set(words))

    original_words = re.findall(r'\b[a-zA-Z]+\b', sentence_string)
    if original_words:
        analysis["longest_word"] = max(original_words, key=len)

    vowels = "aeiou"
    total_vowels = 0
    for word in words:
        for char in word:
            if char in vowels:
                total_vowels += 1
    analysis["total_vowel_count"] = total_vowels

    analysis["is_palindrome"] = is_palindrome(sentence_string)

    return analysis

sentence1 = "I'm kashi."
analysis1 = analyze_sentence(sentence1)
print(f"Sentence: '{sentence1}'\nAnalysis: {analysis1}")

sentence2 = "he is funny."
analysis2 = analyze_sentence(sentence2)
print(f"\nSentence: '{sentence2}'\nAnalysis: {analysis2}")

sentence3 = "how are you "
analysis3 = analyze_sentence(sentence3)
print(f"\nSentence: '{sentence3}'\nAnalysis: {analysis3}")

sentence4 = "i am fine"
analysis4 = analyze_sentence(sentence4)
print(f"\nSentence: '{sentence4}'\nAnalysis: {analysis4}")

sentence5 = ""
analysis5 = analyze_sentence(sentence5)
print(f"\nSentence: '{sentence5}'\nAnalysis: {analysis5}")
```

```python
#18
def get_prime_factors(n):
    if not isinstance(n, int):
        raise TypeError("Input must be an integer.")
    if n <= 0:
        raise ValueError("Input must be a positive integer.")

    factors = []
    d = 2

    while d * d <= n:
        while (n % d) == 0:
            factors.append(d)
            n //= d
        d += 1

    if n > 1:
        factors.append(n)

    return factors

print(f"Prime factors of 12: {get_prime_factors(12)}")
print(f"Prime factors of 30: {get_prime_factors(30)}")
print(f"Prime factors of 100: {get_prime_factors(100)}")
print(f"Prime factors of 7: {get_prime_factors(7)}")
print(f"Prime factors of 1: {get_prime_factors(1)}")
print(f"Prime factors of 999: {get_prime_factors(999)}")

```

```python
#19
import random
from collections import defaultdict

def simulate_dice_rolls(n_times):
    if not isinstance(n_times, int):
        raise TypeError("Input 'n_times' must be an integer.")
    if n_times <= 0:
        raise ValueError("Input 'n_times' must be a positive integer.")
    sum_frequencies = defaultdict(int)

    for _ in range(n_times):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        current_sum = die1 + die2
        sum_frequencies[current_sum] += 1

    return dict(sorted(sum_frequencies.items()))
num_rolls_1 = 100
frequencies_1 = simulate_dice_rolls(num_rolls_1)
print(f"Simulation with {num_rolls_1} rolls: {frequencies_1}")

num_rolls_2 = 1000
frequencies_2 = simulate_dice_rolls(num_rolls_2)
print(f"\nSimulation with {num_rolls_2} rolls: {frequencies_2}")

num_rolls_3 = 100000
frequencies_3 = simulate_dice_rolls(num_rolls_3)
print(f"\nSimulation with {num_rolls_3} rolls: {frequencies_3}")

import matplotlib.pyplot as plt

def plot_distribution(frequencies, title):
    sums = list(frequencies.keys())
    counts = list(frequencies.values())
    total_rolls = sum(counts)
    probabilities = [c / total_rolls for c in counts]

    plt.figure(figsize=(10, 6))
    plt.bar(sums, probabilities, color='skyblue')
    plt.title(title)
    plt.xlabel('Sum of Two Dice')
    plt.ylabel('Probability')
    plt.xticks(range(2, 13))
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

plot_distribution(frequencies_1, f'Probability Distribution for {num_rolls_1} Dice Rolls')
plot_distribution(frequencies_2, f'Probability Distribution for {num_rolls_2} Dice Rolls')
plot_distribution(frequencies_3, f'Probability Distribution for {num_rolls_3} Dice Rolls')

```

```python
#20
def cumulative_sum(numbers_list):
    if not isinstance(numbers_list, list):
        raise TypeError("Input must be a list.")

    if not numbers_list:
        return []

    cumulative_sums = []
    current_sum = 0
    for num in numbers_list:
        if not isinstance(num, (int, float)):
            raise TypeError(f"All elements in the list must be numbers. Found type {type(num)} for value {num}.")
        current_sum += num
        cumulative_sums.append(current_sum)

    return cumulative_sums
list1 = [1, 2, 3]
print(f"Original: {list1}, Cumulative Sum: {cumulative_sum(list1)}")

list2 = [4, -2, 5, 0, 1]
print(f"Original: {list2}, Cumulative Sum: {cumulative_sum(list2)}")

list3 = [10.5, 2.0, -3.5]
print(f"Original: {list3}, Cumulative Sum: {cumulative_sum(list3)}")

empty_list = []
print(f"Original: {empty_list}, Cumulative Sum: {cumulative_sum(empty_list)}")

single_element_list = [7]
print(f"Original: {single_element_list}, Cumulative Sum: {cumulative_sum(single_element_list)}")
```
